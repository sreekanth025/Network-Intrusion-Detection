```python
import pandas as pd
import numpy as np

from sklearn.feature_selection import SelectFpr
from sklearn.feature_selection import chi2
from sklearn.preprocessing import Normalizer
from sklearn import preprocessing
```

```python
isot = pd.read_csv('../data/isot/overall.csv', header = None)
isot.head()
```

```python
isot.tail()
```

```python
isot = isot.drop(0, axis=1)
isot.head()
```

```python
isot[209].value_counts()
```

```python
df_x = isot.drop(209, axis=1)
df_y = isot[209]
```

```python
scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))
d = scaler.fit_transform(df_x)
```

```python
scaled_x = pd.DataFrame(d)
scaled_x.head()
```

```python
def preprocess(df_x, df_y):
    x_normalise = Normalizer().fit(df_x)
    df_x = x_normalise.transform(df_x)
    x_new = SelectFpr(chi2, alpha=0.05).fit_transform(df_x, df_y)
    return x_new
```

```python
x_new = preprocess(scaled_x, df_y)
```

```python
_, n_columns = x_new.shape
print(n_columns)
```

```python
column_names = []

for i in range(n_columns):
    column_names.append('f' + str(i))
```

```python
x_dataframe = pd.DataFrame(data=x_new, columns = column_names)
x_dataframe.head()
```

```python
y_dataframe = pd.DataFrame(df_y.to_numpy(), columns=['class'])
y_dataframe.head()
```

```python
y_dataframe.value_counts()
```

```python
normalized_dataset = pd.concat([x_dataframe, y_dataframe], axis=1)
normalized_dataset.head()
```

```python
normalized_dataset.describe()
```

```python
normalized_dataset.to_csv('../data/isot/overall_normalized.csv', index=False)
```

## Splitting Data into multiple parts for git

```python
df_splits = np.array_split(normalized_dataset, 10)

for df in df_splits:
    print(df.shape)
```

```python
for i, df in enumerate(df_splits):
    file_name = '../data/isot/overall_normalized/split-' + str(i) + '.csv'
    df.to_csv(file_name, index=False)
    print(file_name + ' saved')
```
