```python
import pandas as pd
import numpy as np

from sklearn.feature_selection import SelectFpr
from sklearn.feature_selection import chi2
from sklearn.preprocessing import Normalizer
from sklearn import preprocessing
```

```python
d2 = pd.read_csv('../data/isot/day_wise/d2.csv', header=None)
d3 = pd.read_csv('../data/isot/day_wise/d3.csv', header=None)
d4 = pd.read_csv('../data/isot/day_wise/d4.csv', header=None)
d5 = pd.read_csv('../data/isot/day_wise/d5.csv', header=None)
d6 = pd.read_csv('../data/isot/day_wise/d6.csv', header=None)
```

```python
lis = [d2, d3, d4, d5, d6]
days = [2, 3, 4, 5, 6]
```

```python
for df, day in zip(lis, days):
    df['day'] = day
```

```python
d2.head()
```

```python
d3.head()
```

```python
isot = pd.concat(lis)
print(isot.shape)
isot.head()
```

```python
isot.tail()
```

```python
isot['day'].value_counts()
```

```python
isot = isot.drop(0, axis=1)
isot.head()
```

```python
df_x = isot.drop('day', axis=1).drop(208, axis=1)
df_y = isot[208]
df_day = isot['day']
```

```python
scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))
d = scaler.fit_transform(df_x)
scaled_x = pd.DataFrame(d)
scaled_x.head()
```

```python
def preprocess(df_x, df_y):
    x_normalise = Normalizer().fit(df_x)
    df_x = x_normalise.transform(df_x)
    x_new = SelectFpr(chi2, alpha=0.05).fit_transform(df_x, df_y)
    return x_new

x_new = preprocess(scaled_x, df_y)
_, n_columns = x_new.shape
print(n_columns)
```

```python
column_names = []

for i in range(n_columns):
    column_names.append('f' + str(i))
```

```python
x_dataframe = pd.DataFrame(data=x_new, columns = column_names)
x_dataframe.head()
```

```python
y_dataframe = pd.DataFrame(df_y.to_numpy(), columns=['class'])
y_dataframe.head()
```

```python
y_dataframe.value_counts()
```

```python
df_day.head()
```

```python
df_day.value_counts()
```

```python
day_dataframe = pd.DataFrame(df_day.to_numpy(), columns=['day'])
day_dataframe.head()
```

```python
normalized_dataset = pd.concat([x_dataframe, y_dataframe, day_dataframe], axis=1)
normalized_dataset.head()
```

```python
d2_normalized = normalized_dataset[normalized_dataset['day'] == 2]
d3_normalized = normalized_dataset[normalized_dataset['day'] == 3]
d4_normalized = normalized_dataset[normalized_dataset['day'] == 4]
d5_normalized = normalized_dataset[normalized_dataset['day'] == 5]
d6_normalized = normalized_dataset[normalized_dataset['day'] == 6]
```

```python
norm_lis = [d2_normalized, d3_normalized, d4_normalized, d5_normalized, d6_normalized]
days = [2, 3, 4, 5, 6]
```

```python
for df, i in zip(norm_lis, days):
    folder = '../data/isot/day_wise_normalized/'
    file_name = 'day_' + str(i) + '_normalized.csv'
    df = df.drop('day', axis=1)
    df.to_csv(folder + file_name, index=False)
    print(file_name + ' saved to disk')
```
