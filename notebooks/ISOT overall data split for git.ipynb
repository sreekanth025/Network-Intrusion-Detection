```python
import pandas as pd
import numpy as np
```

```python
isot = pd.read_csv('../data/isot/overall.csv', header = None)
isot.head()
```

```python
df_splits = np.array_split(isot, 10)

for df in df_splits:
    print(df.shape)
```

```python
for i, df in enumerate(df_splits):
    file_name = '../data/isot/overall/split-' + str(i) + '.csv'
    df.to_csv(file_name)
    print(file_name + ' saved')
```
