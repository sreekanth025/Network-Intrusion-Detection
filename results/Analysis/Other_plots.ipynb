```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Average Accuracy vs Number of agents

```python
num_agents = [2, 3, 4, 5, 6, 7, 8, 9, 10]

nsl_acc = [
    0.973011589,
    0.978174603,
    0.964047619,
    0.963888889,
    0.964206349,
    0.971507937,
    0.965079365,
    0.962936508,
    0.958333333
          ]

isot_acc = [
    0.997798915,
    0.997405966,
    0.996109102,
    0.996973865,
    0.996305903,
    0.99650226,
    0.995716756,
    0.996305467,
    0.996699411
]
```

```python
fig, axs = plt.subplots(1, 2, figsize=(13, 3))

nsl_plot = axs[0]
nsl_plot.plot(num_agents, nsl_acc, marker='.', markersize=10)
nsl_plot.set_xlim(1.5, 10.5)
nsl_plot.set_ylim(0.4, 1.1)
nsl_plot.set_xlabel("Number of agents")
nsl_plot.set_ylabel("Avg. accuracy")
nsl_plot.set_title('NSL KDD')
nsl_plot.grid(linestyle='--', linewidth=1)


isot_plot = axs[1]
isot_plot.plot(num_agents, nsl_acc, marker='.', markersize=10)
nsl_plot.set_xlim(1.5, 10.5)
isot_plot.set_ylim(0.4, 1.1)
isot_plot.set_xlabel("Number of agents")
isot_plot.set_ylabel("Avg. accuracy")
isot_plot.set_title('ISOT-CID')
isot_plot.grid(linestyle='--', linewidth=1)

fig.text(0.5, 1, 'Average Accuracy vs Number of agents', ha='center', fontdict={'size': 13})
```

# Equation Plot

```python
x = np.linspace(-0.1, 1.2, 100)
formula = '1 + k*(1-x)*(a**(-x))'

fig, axs = plt.subplots(1, 2, figsize=(15, 4))
ax0 = axs[0]
ax1 = axs[1]

k = 10 
a = 200
y1 = eval(formula)
ax0.plot(x, y1)

k = 10 
a = 5
y2 = eval(formula)
ax0.plot(x, y2)

ax0.axvline(x=0, ls='--', color='brown')
ax0.axhline(y=1, ls='--', color='red')
ax0.plot([1], [1], marker='.', markersize=15, mfc='b', mec='w')
ax0.plot([0], [k+1], marker='.', markersize=15, mfc='b', mec='w')

ax0.legend(['k1, a1', 'k2, a2', 'x=0', 'y=1'])
ax0.text(0.3, 13, 'k1 = k2 and a1 > a2', fontsize = 12)
ax0.text(0.14, 3, 'a1', fontsize = 10)
ax0.text(0.3, 6.2, 'a2', fontsize = 10)
ax0.set_title('Impact of value \'a\'')
ax0.set_xlabel('x')
ax0.set_ylabel('y = attention_multiplier')




k = 30 
a = 20
y1 = eval(formula)
k1 = k
ax1.plot(x, y1)

k = 10 
a = 20
k2 = k
y2 = eval(formula)
ax1.plot(x, y2)

ax1.axvline(x=0, ls='--', color='brown')
ax1.axhline(y=1, ls='--', color='red')
ax1.plot([1], [1], marker='.', markersize=15, mfc='b', mec='w')
ax1.plot([0], [k1+1], marker='.', markersize=15, mfc='b', mec='w')
ax1.plot([0], [k2+1], marker='.', markersize=15, mfc='b', mec='w')


ax1.legend(['k1, a1', 'k2, a2', 'x=0', 'y=1'])
ax1.text(0.35, 30, 'k1 > k2 and a1 = a2', fontsize = 12)
ax1.text(0.05, 30, 'k1', fontsize = 10)
ax1.text(0.05, 10, 'k2', fontsize = 10)
ax1.set_title('Impact of value \'k\'')
ax1.set_xlabel('x')
ax1.set_ylabel('y = attention_multiplier')
```
